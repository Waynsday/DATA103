```python
import pandas as pd 
from sklearn.preprocessing import OneHotEncoder
```

```python
data_df = pd.read_csv("loan_approval.csv")
data_df.head()
```

## <span style="color:blue">Summary</span>

The dataset is a compilation of loan applications that contained borrower's demographics including the loan details and their actual results. The goal of this model report is to compare two models: Random Forest and Support Vector Machine, in the application of predictively classifying a loan application for approval or for rejection. The two models were selected on the basis of their raw computational needs wherein the SVM model is expected to require less time than the RF model. The dataset was pre-processed by imputing all the categorical null values using the mode of the dataset while replacing null numerical values with the dataset's mean. Then, two models were generated: the SVM and RF, where the former was subjected to hyperparameter optimization due to the fact that the SVM is a single model while RF is an ensemble learner. The resulting models showed the SVM outperforming the RF model in accuracy (76.22%), precision (73.46%), and F1 score (84.40%) with only its recall (99.17%) underperforming. In comparison, the RF resulted in an accuracy of 70.27%, a precision of 68.57%, a recall of 100%, and an F1 score of 81.36%. While in this experiment the SVM outperformed the RF model, this loss can be attributed to the lack of parameter optimization with the RF model. Thus, it is recommended that in future model evaluations, the RF model should be subjected to hyperparameter tuning as well. Other recommendations include exploring other algorithms such as Logistic Regression, KNN, neural networks, and the like.

## <span style="color:blue">Data Preprocessing</span>

The data was first inspected for null and missing values. Given the conditions of non-removal, the following columns were imputed using the mode of the dataset: Gender, Married, Dependents, Self_Exmployed, Loan_Amount_Term, and Credit_History. Then the missing Loan Amounts were imputed using the mean of the dataset. Afterwards, binary data were replaced with 1s and 0s and the multi-categorical data were subjected to onehotencoding to be converted to more binary data. The data was split into training and testing with a 70-30 split resulting in 429 observations of training data and 185 observations of testing data.

```python
data_df.isna().sum()
```

```python
data_df['Gender'] = data_df['Gender'].fillna(data_df.Gender.mode()[0])
data_df['Married'] = data_df['Married'].fillna(data_df.Married.mode()[0])
data_df['Dependents'] = data_df['Dependents'].fillna(data_df.Dependents.mode()[0])
data_df['Self_Employed'] = data_df['Self_Employed'].fillna(data_df.Self_Employed.mode()[0])
data_df['LoanAmount'] = data_df['LoanAmount'].fillna(data_df.LoanAmount.mean())
data_df['Loan_Amount_Term'] = data_df['Loan_Amount_Term'].fillna(data_df.Loan_Amount_Term.mode()[0])
data_df['Credit_History'] = data_df['Credit_History'].fillna(data_df.Credit_History.mode()[0])
```

```python
data_df['Education'].value_counts()
```

```python
data_df.Gender = data_df.Gender.apply(lambda x: 1 if x == 'Female' else 0)
data_df.Education = data_df.Education.apply(lambda x: 1 if x == 'Graduate' else 0)
data_df.Loan_Status = data_df.Loan_Status.apply(lambda x: 1 if x == 'Y' else 0)
#convert SeniorCitizen, Partner, Dependents, PhoneService, PaperlessBilling, Churn & etc
for c in ['Married', 'Self_Employed']:
    data_df[c] = data_df[c].apply(lambda x: 1 if x == 'Yes' else 0) 
    
multi_categorical_features = ['Loan_Amount_Term', 'Property_Area', 'Dependents']
enc = OneHotEncoder(handle_unknown='ignore')
res = enc.fit_transform(data_df[multi_categorical_features])
res_df = pd.DataFrame(res.todense(), columns=enc.get_feature_names())

data_df.drop(multi_categorical_features, axis=1, inplace=True)
data_df = pd.concat([data_df, res_df], axis=1)

data_df.head()
```

```python
data_df.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data_df.drop(columns = ['Loan_ID', 'Loan_Status'])
y = data_df['Loan_Status']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print("Train ",X_train.shape)
print("Test ",X_test.shape)
```

## <span style="color:blue">Model Development</span>

There are two models made, one using Support Vector Machines (SVMs) and another using the Random Forest Classifier (RF). These two algorithms were selected in consideration of computational speed but SVM was trained using GridSearch for parameter optimization considering SVM uses only a single model. 

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import  make_scorer
from sklearn.metrics import precision_score, recall_score, f1_score, r2_score, accuracy_score
from sklearn.svm import SVC
```

```python
param_grid = {
    'C': [0.1, 0.001, 0.0001, 1, 10, 100],
    'kernel': ['rbf','sigmoid', 'linear']
}

scorers = {
    'precision_score': make_scorer(precision_score),
    'recall_score': make_scorer(recall_score),
    'accuracy_score': make_scorer(accuracy_score)
}
```

```python
model = SVC()
search = GridSearchCV(model, param_grid, verbose=3, cv =10)
search.fit(X_train, y_train)
```

```python
search.best_estimator_
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth =2)
clf.fit(X_train, y_train)
```

```python
best_search = search.best_estimator_
y_pred = best_search.predict(X_test)
print("Accuracy: {:.2f}%".format(accuracy_score(y_test, y_pred)*100))
print("Precision: {:.2f}%".format(precision_score(y_test, y_pred)*100))
print("Recall: {:.2f}%".format(recall_score(y_test, y_pred)*100))
print("F1-Score: {:.2f}%".format(f1_score(y_test, y_pred)*100))
```

```python
y2_pred = clf.predict(X_test)
print("Accuracy: {:.2f}%".format(accuracy_score(y_test, y2_pred)*100))
print("Precision: {:.2f}%".format(precision_score(y_test, y2_pred)*100))
print("Recall: {:.2f}%".format(recall_score(y_test, y2_pred)*100))
print("F1-Score: {:.2f}%".format(f1_score(y_test, y2_pred)*100))
```

## <span style="color:blue">Model Interpretability</span>

Explaining each model starting with the SVM, the SVM model highlights credit history to be largest indicator of loan rejection. In terms of location, semi-urban and rural locations indicate likelihood of loan approval surprisingly. The loan amount plays a significant role in the SVM model as well with that variable showing higher loan amounts to incidate higher likelihood of approval and lower loan amounts indicate a lower likelihood of approval. Intuitively, it makes sense given the higher profitability of large loans rather than smaller loans. 

On the other hand, the RF model similarly holds credit history to be the most important feature indicating a high likelihood of rejection given the lack of credit history. The rural and semi-urban residences also similarly significantly affect the likelihood of a loan approval like the SVM model. In this model, loan terms are ranked higher up in feature importance compared to the SVM model.
